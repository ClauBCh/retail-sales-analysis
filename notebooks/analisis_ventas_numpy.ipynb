### Carga y Preprocesamiento de Datos:

```python
import numpy as np

def cargar_datos(ruta_archivo):
    # Carga los datos del archivo CSV utilizando NumPy como texto
    datos = np.genfromtxt(
        ruta_archivo,
        delimiter=',',
        skip_header=1,
        dtype=str,
        encoding='utf8'
    )
    return datos

def cargar_columnas(ruta_archivo):
    # Carga los nombres de las columnas
    columnas = np.genfromtxt(
        ruta_archivo,
        delimiter=',',
        max_rows=1,
        dtype=str,
        encoding='utf8'
    )
    return columnas

    #Carga del CVS con NumPy
if __name__ == "__main__":
    ruta_archivo = '/content/drive/MyDrive/Colab Notebooks/retail_sales.csv'
    columnas = cargar_columnas(ruta_archivo)
    datos = cargar_datos(ruta_archivo)

  # Exploración del archivo
    print("Columnas")
    print(columnas)

    print("\nHEAD - Primeras 5 filas:")
    print(datos[:5])

    print("\nTAIL - Úlitmas 5 filas:")
    print(datos[-5:])

    print("\nSHAPE - Dimensiones del dataset:")
    print(datos.shape)

  #Información básica del dataset
    print("\nInformación básica del dataset:")
    print("Cantidad de filas:", datos.shape[0])
    print("Cantidad de columnas:", datos.shape[1])
    print("Tipo de datos del array:", datos.dtype)
```

## Exploración de datos

```python
# Selección de columnas relevantes para el análisis
categorias = datos[:, 5]
cantidad = datos[:, 6].astype(int)
precio_unitario = datos[:, 7].astype(float)
monto_total = datos[:, 8].astype(float)

# Revisión de categorías únicas
categorias_unicas = np.unique(categorias)

print("Categorías únicas:")
print(categorias_unicas)

print("\nTotal general de ventas:")
print(np.sum(monto_total))

print("\nPromedio general de ventas:")
print(np.mean(monto_total))

print("\nVenta mínima:")
print(np.min(monto_total))

print("\nVenta máxima:")
print(np.max(monto_total))
```

### Manipulación de Datos

```python
#Calcular el total de ventas por categoría única de producto.
print("Total de ventas por categoría:")
total_por_categoria = np.zeros(len(categorias_unicas))


# Total ventas por categoría
for i, categoria in enumerate(categorias_unicas):
    total_categoria = np.sum(monto_total[categorias == categoria])
    total_por_categoria[i] = total_categoria
    print(f"{categoria}: {total_categoria}")
```

```python
#Calcular el promedio de ventas diarias por categoría de producto.
print("Promedio de ventas por categoría:")
promedio_por_categoria = np.zeros(len(categorias_unicas))

#Promedio de ventas por categoría
for i, categoria in enumerate(categorias_unicas):
    promedio_categoria = np.mean(monto_total[categorias == categoria])
    promedio_por_categoria[i] = promedio_categoria
    print(f"{categoria}: {promedio_categoria:.2f}")
```

```python
#Calcular las categorías con mayor y menor ventas
indice_mayor_venta = np.argmax(total_por_categoria)
indice_menor_venta = np.argmin(total_por_categoria)

categoria_mayor = categorias_unicas[indice_mayor_venta]
categoria_menor = categorias_unicas[indice_menor_venta]

mayor_venta = total_por_categoria[indice_mayor_venta]
menor_venta = total_por_categoria[indice_menor_venta]

# Mayor venta
print("Categoría con mayor venta:")
print(f"{categoria_mayor}: {mayor_venta}")

#Menor venta
print("\nCategoría con menor venta:")
print(f"{categoria_menor}: {menor_venta}")
```

```python
# Promedio de ventas diarias por categoría
fechas = datos [:, 1]
print("Promedio de ventas diarias por categoría:")

for categoria in categorias_unicas:
    filtro_categoria = categorias == categoria
    fechas_categoria = fechas[filtro_categoria]
    ventas_categoria = monto_total[filtro_categoria]

    fechas_unicas = np.unique(fechas_categoria)
    total_dias = len(fechas_unicas)

    promedio_diario = np.sum(ventas_categoria) / total_dias

    print(f"{categoria}: {promedio_diario:.2f}")
```
